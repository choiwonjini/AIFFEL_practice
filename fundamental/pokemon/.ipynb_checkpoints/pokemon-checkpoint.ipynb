# 3. 전처리

## 3-1. 이름

### 3-1-1. 이름의 길이

```python
# name_count 컬럼을 생성
pokemon["name_count"] = pokemon['Name'].apply(lambda i: len(i))

# 길이가 10 이상이면 True, 미만이면 False를 가지는 long_name 컬럼 생성
pokemon["long_name"] = pokemon['name_count'] >= 10
```

### 3-1-2. 이름에 자주 쓰이는 토큰 추출

포켓몬의 이름은 총 네 가지 타입으로 나뉜다.  

* 한 단어 ex. Venusaur
* 두 단어이고, 앞 단어는 두 개의 대문자를 가지며 대문자를 기준으로 두 부분으로 나뉘는 경우 ex. VenusaurMega Venusaur
* 이름은 두 단어이고, 맨 뒤에 X, Y로 성별을 표시하는 경우 ex. CharizardMega Charizard X
* 알파벳이 아닌 문자를 포함하는 경우 ex. Zygarde50% Forme

(1) 이름에 알파벳이 아닌 문자가 들어간 경우 전처리

```python
# isalpha() 함수 사용을 위해 일단 공백을 없앤 임시 칼럼 생성
pokemon["Name_nospace"] = pokemon["Name"].apply(lambda i: i.replace(" ", ""))

# 알파벳으로만 이루어진 이름인지 확인하는 칼럼 생성
pokemon['name_isalpha'] = pokemon["Name_nospace"].apply(lambda i: i.isalpha())
pokemon.head()
```

```python
print(pokemon[pokemon["name_isalpha"]==False].shape)
pokemon[pokemon["name_isalpha"]==False]
```

9마리 뿐이니 적당히 바꿔주기

```python
pokemon = pokemon.replace(to_replace="Nidoran♀", value="Nidoran X")
pokemon = pokemon.replace(to_replace="Nidoran♂", value="Nidoran Y")
pokemon = pokemon.replace(to_replace="Farfetch'd", value="Farfetchd")
pokemon = pokemon.replace(to_replace="Mr. Mime", value="Mr Mime")
pokemon = pokemon.replace(to_replace="Porygon2", value="Porygon Two")
pokemon = pokemon.replace(to_replace="Ho-oh", value="Ho Oh")
pokemon = pokemon.replace(to_replace="Mime Jr.", value="Mime Jr")
pokemon = pokemon.replace(to_replace="Porygon-Z", value="Porygon Z")
pokemon = pokemon.replace(to_replace="Zygarde50% Forme", value="Zygarde Forme")

pokemon.loc[[34, 37, 90, 131, 252, 270, 487, 525, 794]]
```

```python
# 다시 확인
pokemon["Name_nospace"] = pokemon['Name'].apply(lambda i: i.replace(" ", ""))
pokemon["name_isalpha"] = pokemon["Name_nospace"].apply(lambda i: i.isalpha())

pokemon[pokemon["name_isalpha"]==False]
```

(2) 이름을 띄어쓰기 & 대문자 기준으로 분리해 토큰화하기

```python
import re

def tokenize(name):
    name_split = name.split() # 공백 기준으로 분리
    tokens = []
    for part_name in name_split:
        a = re.findall("[A-Z][a-z]*", part_name) # 대문자로 시작해서 소문자로 끝나는 문자 추출
        tokens.extend(a) # 토큰 리스트에 추가

    return np.array(tokens)
```

```python
# 전설 포켓몬의 토큰
all_tokens = list(legendary["Name"].apply(tokenize).values) # 여러 개의 리스트가 합쳐진 형태

token_set = []
for token in all_tokens:
    token_set.extend(token) # 모든 토큰들을 하나의 리스트로 합치기

print(len(set(token_set)))
print(token_set)
```

```python
# 빈번하게 나온 토큰들 찾기
from collections import Counter

most_common = Counter(token_set).most_common(10)
most_common
```

```python
# 전설의 포켓몬 이름에 등장하는 토큰 상위 10개가 포켓몬의 이름에 있는지의 여부를 나타내는 컬럼들 생성
for token, _ in most_common:
    # pokemon[token] = ... 형식으로 사용하면 뒤에서 warning이 발생
    pokemon[f"{token}"] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

# 4. 모델링

## 4-1. 베이스라인(Baseline) 모델
* 가장 기초적인 방법으로 만든 모델  
* 성능은 안 좋을지 모르지만, 성능 하한선을 제공함으로써 우리가 새롭게 만들 모델이 맞는 방향으로 가고 있는지 확인할 수 있게 도와준다.

원본 데이터로 모델링

```python
original_data.columns
```

모델 학습을 위해선 문제열 칼럼, target을 제외해야 함

```python
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
target = 'Legendary'

X = original_data[features]

print(X.shape)
X.head()
```

```python
y = original_data[target]

print(y.shape)
y.head()
```

```python
# train, test 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

decision tree 모델 학습

```python
%%time
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

model = DecisionTreeClassifier(random_state=25)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

Positive는 Legendary=True(1)  
Negative는 Legendary=False(0)  

Accuracy = (144+8)/(144+3+5+8) = 0.95  
정확도가 95%지만 불균형 데이터이기 때문에 해석에 주의해야 한다.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

recall = 0.62로 꽤 낮다.  
-> FN 값이 높음 (recall = TP/(TP+FN))  
-> 전설의 포켓몬인데 일반 포켓몬이라고 판단하는 경우가 많다.
