# 2. L2 Regularization (Ridge)

```python
# L2 정칙화 - Ridge
for x, y in train_dataloader:
    x = x.to(device)
    y = y.to(device)

    output = model(x)

    _lambda = 0.5
    # 가중치 제곱의 합 사용
    l2_loss = sum(p.pow().sum() for p in model.parameters())

    loss = criterion(output, y) + _lambda * l2_loss
```
