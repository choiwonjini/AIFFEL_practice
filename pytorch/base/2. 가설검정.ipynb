# 가설검정

쌍체 t-검정  
- 동일 집단에 대한 학습 방법 전후 효과 검정  

비쌍체 t-검정
- 서로 다른 두 개의 독립적인 집단(실험군, 대조군) 간에 유의미한 차이가 있는지 검정  
<br>

ML의 통계적 가설을 적용한다면 비쌍체 t-검정을 사용한다.  
- X, Y 사이에 유의한 차이가 있는지 검정하는데, 이대 변수들의 샘플 데이터는 독립항등분포를 따르기 때문.

```python
# 성별에 따른 키 차이 검정
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

male_height = stats.norm.rvs(loc=170, scale=10, size=500, random_state = 1)
female_height = stats.norm.rvs(loc=150, scale=10, size=500, random_state = 1)

X = np.concatenate([male_height, female_height])
Y = ['male'] * len(male_height) + ['female'] * len(female_height)

df = pd.DataFrame(list(zip(X, Y)), columns=['X', 'Y'])
fig = sns.displot(data=df, x='X', hue='Y', kind='kde')
fig.set_axis_labels('cm', 'count')
plt.show()
```

```python
# 비쌍체 t-검정
statistic, pvalue = stats.ttest_ind(male_height, female_height, equal_var=True)

statistic, pvalue
```

두 그룹은 유의한 차이가 있다.
